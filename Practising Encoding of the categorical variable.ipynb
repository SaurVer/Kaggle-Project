```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
```

```python
data=pd.read_csv('train.csv')
```

```python
catdata=data.select_dtypes(include=object)
```

```python
catdata.head(5)
```

```python
catdata.dropna(thresh=.4*len(catdata),axis=1,inplace=True)
```

```python
catdata.isnull().sum()
```

```python
# just to practise I will straightaway move on to encoding if all categories.
```

```python
#Creating a dictionary
import warnings
warnings.filterwarnings('ignore')
encoderdict={}
for item in list(catdata.columns):
    item_not_null=catdata[catdata[item].notnull()][item]
    encoderdict[item]= OrdinalEncoder()
    item_not_null=item_not_null.values.reshape(-1,1)
    encoded_value=encoderdict[item].fit_transform(item_not_null)
    # Putting it into the daafram catdataabs
    catdata.loc[catdata[item].notnull(),item]=np.squeeze(encoded_value)
```
